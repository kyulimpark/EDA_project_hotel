```python
# 로지스틱 회귀분석
# 참고 출처1: http://hleecaster.com/ml-logistic-regression-example/
# 참고 출처2: https://myjamong.tistory.com/80
import pickle
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df2.columns
```

```python
# 독립변수, 종속변수 설정
x = df2[["previous_cancellations", "lead_time", "total_of_special_requests"]]
y = df2["is_canceled"]
```

```python
# 학습 데이터, 테스트 데이터 설정
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
# 데이터 스케일링
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)
```

```python
# 로지스틱회귀분석모델
model = LogisticRegression(solver = "liblinear")
model.fit(train_x, train_y)
```

```python
print(model.score(train_x, train_y)) # 학습데이터 정확도
```

```python
print(model.score(test_x, test_y)) # 테스트데이터 정확도
```

```python
print(model.coef_) #독립변수별 계수, 독립변수가 종속변수에 미치는 영향도
```
